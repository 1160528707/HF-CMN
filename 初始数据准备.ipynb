```python
from lxml import etree
import json
import os
from collections import Counter
import random



def dict_save_json(dict, file_name):
    with open('{}'.format(file_name), 'w') as file_position:
        json.dump(dict, file_position)
    print('save json file: {}'.format(file_name))


if __name__ == '__main__':
    frontal_image_path = 'D:\\Xinshuai\\mimic\\Tre_total_CXR\\'
    report_path='D:\\Xinshuai\\mimic\\Tre_total_Report_tre_Last\\Cardiomegaly\\'
    # The image name
    frontal_names_lst = []
    # The image number after CXR
    frontal_num_lst = []
    report_name_lst=[]
    report_con=[]
    # one_sentence_impression = False
    json_dict = {}
    json_dict_train = {}
    json_dict_test = {}
    # The image dataset with only frontal views is selected manually.
    for frontal_image_name in os.listdir(frontal_image_path):
        frontal_names_lst.append(str(frontal_image_name).split(".")[0])

    for name in os.listdir(report_path):
        xml_file = os.path.join(report_path, name)
        report_name_lst.append(str(name).split(".")[0])
        with open(xml_file,'r') as report:
            report_con.append(report.readlines())

    for image_name in range(0,len(frontal_names_lst)):
        for report_name in range(0,len(report_name_lst)):
            if report_name_lst[report_name]==frontal_names_lst[image_name]:
                json_dict[frontal_names_lst[image_name]] = report_con[image_name]
                if random.random() < 0.3 and len(json_dict_test.keys()) < 3000:
                    json_dict_test[frontal_names_lst[image_name]] = report_con[image_name]
                else:
                    json_dict_train[frontal_names_lst[image_name]] = report_con[image_name]
    
    if len(json_dict_test.keys()) != 3000:
        print("Please run this program AGAIN!")
    assert len(json_dict_test.keys()) == 3000

    print('The number of complete radiology reports with the frontal view in total:', len(json_dict.keys()))
    print('The number of complete radiology reports with the frontal view in training set:', len(json_dict_train.keys()))
    print('The number of complete radiology reports with the frontal view in testing set:', len(json_dict_test.keys()))

    # gnerate json file from the dictionary
    dict_save_json(json_dict, "D:\\Xinshuai\\mimic\\json\\C\\C_total.json")
    dict_save_json(json_dict_train, "D:\\Xinshuai\\mimic\\json\\C\\C_total_train.json")
    dict_save_json(json_dict_test, "D:\\Xinshuai\\mimic\\json\\C\\C_total_test.json")
```
